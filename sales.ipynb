Create Shrinkage(missing products due to admin fault,misplace,theft,counting issues) Metric :

Formula Hint: shrinkage = units_ordered - units_sold - inventory_level(actual product left in inventory

eg 100-80-15=5 shirnkage).

```python
df['shrinkage']= df['Units Ordered'] - df['Units Sold'] - df['Inventory Level']
```

calculate forecast error and accuracy

forecast_error = abs(forecast - units_sold)

forecast_accuracy = 1 - (forecast_error / forecast)

```python
df['forecast_error']=( df['Demand Forecast'] - df['Units Sold']).abs()
```

```python
df['forecast_accuracy'] = 1 - (df['forecast_error'] / df['Demand Forecast'])
```

Explore Price vs. Units Sold(hexbinplot)

```python
df['Units Sold'].describe()
```

```python
plt.figure(figsize=(10,6))
plt.hexbin(df['Price'], df['Units Sold'], gridsize=30, cmap='Blues')
plt.colorbar(label='Count in Bin')
plt.xlabel("Price")
plt.ylabel("Units Sold")
plt.title("Hexbin Plot: Price vs Units Sold")
plt.tight_layout()
plt.show()
# as scatter plot is not suitable for large datasets(it becomes overcroweded), we use hexbin plot to visualize the relationship between price and units sold.
```

Average Units Sold vs Price Range(barplot)

```python
df['price_bin'] = pd.cut(df['Price'], bins=5)  # Group prices into 5 bins
grouped = df.groupby('price_bin')['Units Sold'].mean().reset_index()


plt.figure(figsize=(10,5))
sns.barplot(x='price_bin', y='Units Sold', data=grouped)
plt.xticks(rotation=45)
plt.title("Average Units Sold by Price Range")
plt.tight_layout()
plt.show()
```

units ordered vs units sold(scatterplot)

```python
dfsample_head=df.head(500)
plt.figure(figsize=(10,6))
sns.scatterplot(x='Units Ordered', y='Units Sold', data= dfsample_head)
plt.title("units ordered vs units sold at start of dataset")
plt.xlabel("units ordered")
plt.ylabel("units sold")
plt.show()
```

```python
dfsample_tail=df.tail(500)
plt.figure(figsize=(10,6))
sns.scatterplot(x='Units Ordered',y='Units Sold', data =dfsample_tail)
plt.title("units ordered vs units sold at end of dataset")
plt.xlabel("units ordered")
plt.ylabel("units sold")
plt.show()
```

Box Plot by Price Bins

```python
df['pricebin']=pd.cut(df['Price'],bins=3)
plt.figure(figsize=(10,6))
sns.boxplot(x='pricebin', y='Units Sold', data= df)
plt.title("box plot by price")
plt.xlabel("price_bin")
plt.ylabel("units sold")
plt.show()
```

histogram of units sold

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Units Sold'],bins=15, kde=True)
plt.title("histogram of unit sold")
plt.xlabel("units sold")
plt.ylabel("frequency")
plt.show()
```
